```python
import os, sys, re
from xlrd import open_workbook
from pprint import pprint
```

```python
worksheet_dir = "/Volumes/DFS_Chaussabel_LabShare/Illumina HiScan SQ/Cluster Worksheets/"
```

```python
worksheets_2015 = [ w for w in os.listdir(worksheet_dir) if re.search('^2015', w) ]
```

```python
worksheet = [ os.path.join(worksheet_dir, w) for w in worksheets_2015 if 'Oct' in w ][0]
```

```python
for w in worksheets_2015:
    if re.search('xlsx$', w):
        print w
        worksheet = os.path.join(worksheet_dir, w)
        book = open_workbook(worksheet)

        lib_pool_sheet = [ s for s in book.sheet_names() if 'lib' in s.lower() ]
        if len(lib_pool_sheet):
            lib_pool_sheet = lib_pool_sheet[0]

            sheet = book.sheet_by_name(lib_pool_sheet)
            keys = [ sheet.cell(0, col_index).value \
                     for col_index in xrange(sheet.ncols) ]
            print keys
            keys_dict = {'lib.*id': 'lib_id',
                         'project': 'project',
                         'lane': 'lane',
                         'type': 'type'}
            lib_dict = {}
            for row_idx in xrange(1, sheet.nrows):
                d = { keys_dict[k]: sheet.cell(row_idx, col_idx).value \
                      for col_idx in xrange(sheet.ncols) \
                      for k in keys_dict \
                      if re.search(k, keys[col_idx].lower()) }
                lib = d['lib_id']
                del d['lib_id']
                lib_dict[lib] = d
                
            pprint(lib_dict)

        else:
            print "lib info not available"

        
```
